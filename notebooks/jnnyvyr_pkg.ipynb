```python
# Add src to path using absolute path
import sys
import os

# Get the absolute path to src folder
notebook_dir = os.path.dirname(os.path.abspath('__file__'))
src_path = os.path.join(notebook_dir, '..', 'src')
sys.path.insert(0, os.path.abspath(src_path))

# Now import
from preprocessing import (
    chunkify_with_time,
    generate_target_labels_aligned,
    extract_combined_features_sequential,
    train_test_split_chunks_stratified
)

# Data Loading (stays in notebook)
import pandas as pd
file_id = '1URua1BRmcTgkwU1nGPWRkrxhabWVotxH'
url = f'https://drive.google.com/uc?id={file_id}&export=download'
df_tochunk = pd.read_feather(url)

file_id = '1qGfSIb9EEJ4ZxlWnBcsILgh9LbHAiMld'
url = f'https://drive.google.com/uc?id={file_id}&export=download'
gl_d1 = pd.read_feather(url)

# Use package functions
CHUNK_SIZE = 60 * 60 * 250
STEP_SIZE = 60 * 250

chunks_with_time = chunkify_with_time(df_tochunk, CHUNK_SIZE, STEP_SIZE)
target_labels = generate_target_labels_aligned(chunks_with_time, gl_d1)

aligned_chunks = [chunk for chunk, _, _ in chunks_with_time]
X_train, X_test, y_train, y_test = train_test_split_chunks_stratified(
    aligned_chunks, target_labels, test_size=0.2
)

X_train_features = extract_combined_features_sequential(X_train)
X_test_features = extract_combined_features_sequential(X_test)

# Model training stays in notebook...
```

### Model Training

```python
# Handle NaN values - fill with column means or drop rows
print(f"NaN values in X_train: {X_train_features.isna().sum().sum()}")
print(f"NaN values in X_test: {X_test_features.isna().sum().sum()}")
```

```python
X_train_features = X_train_features.fillna(X_train_features.mean())
X_test_features = X_test_features.fillna(X_train_features.mean())  # Use train means for test
```

```python
# Handle NaN values - fill with column means or drop rows
print(f"NaN values in X_train: {X_train_features.isna().sum().sum()}")
print(f"NaN values in X_test: {X_test_features.isna().sum().sum()}")
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


#Define and train an SVM model
svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=10.0, gamma="scale", probability=True))
])
svm_clf.fit(X_train_features, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, auc

# Evaluate the model using Precision-Recall AUC
y_train_pred_proba = svm_clf.predict_proba(X_train_features)[:, 1]
precision, recall, _ = precision_recall_curve(y_train, y_train_pred_proba)
pr_auc_train = auc(recall, precision)
print(f"Scoring PR-AUC on y_train: {pr_auc_train:.4f}")

y_test_pred_proba = svm_clf.predict_proba(X_test_features)[:, 1]
precision, recall, _ = precision_recall_curve(y_test, y_test_pred_proba)
pr_auc_test = auc(recall, precision)
print(f"Scoring PR-AUC on y_test: {pr_auc_test:.4f}")
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

# Predict on test data
y_test_pred = svm_clf.predict(X_test_features)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {accuracy:.4f}")

# Generate classification report
print(classification_report(y_test, y_test_pred))
```

### Save trained model

```python
# Save the trained SVM model
import os
import pickle


# Ensure the directory exists
model_dir = '../models'
os.makedirs(model_dir, exist_ok=True)

# Save the model
with open(os.path.join(model_dir, 'jnnyvyr_svm_model03.pkl'), 'wb') as f:
    pickle.dump(svm_clf, f)
```
